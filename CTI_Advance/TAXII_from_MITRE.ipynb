```python
from taxii2client.v20 import Server # only specify v20 if your installed version is >= 2.0.0

server = Server("https://cti-taxii.mitre.org/taxii/")
api_root = server.api_roots[0]
# Print name and ID of all ATT&CK domains available as collections
for collection in api_root.collections:
    print(collection.title.ljust(20) + collection.id)
```

```python
enterprise_attack_collection = None
for collection in api_root.collections:
    if "Enterprise ATT&CK" in collection.title:
        enterprise_attack_collection = collection
        break
```

```python
from taxii2client.v20 import Collection

if enterprise_attack_collection:
    collection = Collection(f"{api_root.url}/collections/{enterprise_attack_collection.id}/")
    stix_objects = collection.get_objects()
    for obj in stix_objects["objects"][:10]:
        print(obj)
else:
    print("Enterprise ATT&CK collection not found.")
```

```python
if enterprise_attack_collection:
    collection = Collection(f"{api_root.url}/collections/{enterprise_attack_collection.id}/")
    stix_objects = collection.get_objects()

    procedures = [obj for obj in stix_objects["objects"] if obj["type"] == "attack-pattern"]

    print("\n前 10 條 procedures:")
    for procedure in procedures[:10]:
        print(procedure)
```
